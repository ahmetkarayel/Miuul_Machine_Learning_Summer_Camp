```python
# Cinsiyete göre yaş ortalamasını, max ve minimum yaşı bulalım.

data.groupby("sex").age.agg(["mean", "max", "min"]) # Kolay yol
# data.groupby("sex").agg({"age":["mean", "max", "min"]}) --> Alternatif yol
```

```python
# Hem cinsiyet hem de liman'a göre gruplama yaparak yaş ve survived ortalamasını alalım

data.groupby(["sex", "embark_town"]).agg({"age":["mean"],
                        "survived":["mean"]})
```

### Apply & Lambda

```python
import pandas as pd
import seaborn as sns
pd.set_option("display.max_columns", None)
```

```python
data = sns.load_dataset("titanic")
data.head()
```

---
* **.apply():** Satır yada sütunlarda otomatik olarak fonksiyon çalıştırmaya imkan sağlar. 
* **.lambda():** Bir fonksiyon tanımlama şeklidir. Kullan at fonksiyonlardır. 
***

```python
data["age2"] = data.age*2
data["age3"] = data.age*5
data.head()
```

```python
# İçerisinde "age" geçen tüm kolonları 10'a bölelim

data[["age", "age2", "age3"]].apply(lambda x: x/10).head() # Buradaki x her bir değişkeni temsil ediyor.
```

```python
# İçerisinde "age" geçen tüm kolonları 10'a bölüp ekrana yazdır

data.loc[:, data.columns.str.contains("age")].apply(lambda x: x/10).head()
```

```python
# Normalization (Scaling) işlemi

data.loc[:, data.columns.str.contains("age")].apply(lambda x: (x-x.mean()) / x.std()).head()
```

```python
# "apply" ile kendi tanımladığımız fonksiyonu da kullanabiliriz.

def standart_scaler(col_name):
    return (col_name - col_name.mean()) / col_name.std()

data.loc[:, data.columns.str.contains("age")].apply(standart_scaler).head()
```

```python
# Değişiklikleri kalıcı hale getirmek için

data.loc[:, data.columns.str.contains("age")] = data.loc[:, data.columns.str.contains("age")].apply(standart_scaler).head()
data.head()
```

```python
# merge ile birleştirme işlemleri

data1 = pd.DataFrame({"employees":["john", "dennis", "mark", "maria"],
                     "group": ["accounting", "engineering", "engineering", "hr"]})

data2 = pd.DataFrame({"employees":["mark", "john", "dennis", "maria"],
                     "start_date": [2010, 2009, 2014, 2019]})

```

```python
# employees kolonu üzerinden birleştirme yapalım.

data3 = pd.merge(data1, data2) # Direkt olarak ortak kolonlar üzerinden birleştirme yaptı. Parametre vermedik.
# default on="employees"
data3.head()
```

```python
# Her çalışanın müdürünün bilgisine erişmek istiyoruz.

data4 = pd.DataFrame({"group": ["accounting", "engineering", "hr"],
                     "manager": ["Caner", "Mustafa", "Berkcan"]})

pd.merge(data3, data4, on="group")
```

```python
# Kategorik değişkenler için fonksiyon

def cat_summary(dataframe, col_name):
    print(pd.DataFrame({col_name: dataframe[col_name].value_counts(),
                       "Ratio":100*dataframe[col_name].value_counts() / len(dataframe)}))
    print("##################################")

for col in cat_cols:
    cat_summary(data, col)
```

```python
# Kategorik değişkenler için grafik ile birlikte fonksiyon

def cat_summary(dataframe, col_name, plot=False):
    print(pd.DataFrame({col_name: dataframe[col_name].value_counts(),
                       "Ratio":100*dataframe[col_name].value_counts() / len(dataframe)}))
    print("##################################")
    
    if plot:
        sns.countplot(x=dataframe[col_name], data=dataframe)
        plt.show(block=True) # Birden fazla döngüye girip grafik oluşturacağımız için yazdık.

for col in cat_cols:
    if data[col].dtypes == "bool":
        print(col)
    else:
        cat_summary(data, col, True)
```

```python
cat_cols
```

```python
# Bool tipteki değişkenleri kullanabilmek için "tip dönüşümü" yapıyoruz.

for col in cat_cols:
    if data[col].dtypes == "bool":
        data[col] = data[col].astype(int) # Tip dönüşümü
        cat_summary(data, col, True)
    else:
        cat_summary(data, col, True)
```

### Değişkenlerin Yakalanması ve İşlemlerin Genelleştirilmesi

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
pd.set_option("display.max_columns", None)
```

```python
data = sns.load_dataset("titanic")
data.head()
```

```python
data.info()
```

```python
def grab_col_names(dataframe, cat_th=10, car_th=30):
    """
    Veri setindeki kategorik, numerik ve kategorik fakat kardinal değişkenlerin isimlerini verir.
    
    Parameters
    ----------
    dataframe: dataframe
        değişken isimleri alınmak istenen dataframe'dir.
    cat_th: int, float
        numerik fakat kategorik olan değişkenler için sınıf eşik değeri
    car_th: int, float
        kategorik fakat kardinal değişkenler için sınıf eşik değeri
    
    Returns
    -------
    cat_cols: list
            Kategorik değişken listesi
    num_cols: list
            Numerik değişken listesi
    cat_but_car: list
            Kategorik görünümlü kardinal değişken listesi
            
    Notes
    -----
    cat_cols + num_cols + cat_but_car = toplam değişken sayısı
    num_but_cat cat_cols'un içerisinde.
    Return olan 3 liste toplamı toplam değişken sayısına eşittir.
    
    """
    cat_cols = [col for col in data.columns if str(data[col].dtypes) in ["category", "object", "bool"]]
    num_but_cat = [col for col in data.columns if data[col].dtypes in ["int64", "float64"] and data[col].nunique() < 10]
    cat_but_car = [col for col in data.columns if data[col].nunique() > 20 and str(data[col].dtypes) in ["category", "object"]]
    cat_cols = cat_cols + num_but_cat
    cat_cols = [col for col in cat_cols if col not in cat_but_car]
    
    
    num_cols = [col for col in data.columns if data[col].dtypes in ["int64", "float64"]]
    num_cols = [col for col in num_cols if col not in cat_cols]
    
    print(f"Observations:{dataframe.shape[0]}")
    print(f"Variables:{dataframe.shape[1]}")
    print(f"cat_cols:{len(cat_cols)}")
    print(f"num_cols:{len(num_cols)}")
    print(f"cat_but_car:{len(cat_but_car)}")
    print(f"num_but_cat:{len(num_but_cat)}")
    
    return cat_cols, num_cols, cat_but_car

cat_cols, num_cols, cat_but_car = grab_col_names(data)
```

```python
# Kategorik değişkenler vs Bağımlı Değişkenimiz

def target_summary_with_cat(dataframe, target, categorical_col):
    print(pd.DataFrame({"TARGET_MEAN": dataframe.groupby(categorical_col)[target].mean()}))

for col in cat_cols:
    target_summary_with_cat(data, "survived", col)
    print("#"*20)
```

#### Hedef Değişkenin Sayısa Değişkenler ile Analizi

```python
# Korelasyon 

corr = data[num_cols].corr()
corr
```

#### Yüksek Korelasyonlu Değişkenlerin Silinmesi

```python
cor_matrix = data.corr().abs() # Mutlak değer alınıyor.
```

```python
# Heatmap üzerinden birbirini tekrar edecek değerlerden kurtuluyoruz.

upper_triangle_matrix = cor_matrix.where(np.triu(np.ones(cor_matrix.shape), k=1).astype(np.bool))

# Birbiriyle yüksek korelasyonlu değerler

drop_list = [col for col in upper_triangle_matrix.columns if any(upper_triangle_matrix[col] > 0.90)] # 0.9 th
drop_list
```

```python
# Yüksek korelasyonlulardan kurtulmak

data.drop(drop_list, axis=1)
```

```python
# Fonksiyonlaştırma

def high_correlated_cols(dataframe, plot=False, corr_th=0.90):
    corr = dataframe.corr()
    cor_matrix = corr.abs()
    upper_triangle_matrix = cor_matrix.where(np.triu(np.ones(cor_matrix.shape), k=1).astype(bool))
    drop_list = [col for col in upper_triangle_matrix.columns if any(upper_triangle_matrix[col] > corr_th)]
    if plot:
        import seaborn as sns
        import matplotlib.pyplot as plt
        sns.set(rc={'figure.figsize': (15, 15)})
        sns.heatmap(corr, cmap="RdBu")
        plt.show()
    return drop_list


# high_correlated_cols(df)
drop_list = high_correlated_cols(data, plot=True)
data.drop(drop_list, axis=1)
high_correlated_cols(data.drop(drop_list, axis=1), plot=True)
```
